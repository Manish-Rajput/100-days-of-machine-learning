MAE, MSE & RMSE are loss functions. They are not absolute and depend on dataset. i.e. 1.5 loss in case of placement data (where target variable is package in lpa) represents loss of 1.5 lpa in case MAE. in case of MSE unit will be in squared

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
df.shape
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
y_pred = lr.predict(X_test)
```

```python
y_test.values
```

```python
print("MAE",mean_absolute_error(y_test,y_pred))
```

```python
print("MSE",mean_squared_error(y_test,y_pred))
```

```python
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
print("R2 score",r2_score(y_test,y_pred))
r2 = r2_score(y_test,y_pred)
```

```python
# Adjusted R2 score
X_test.shape
```

```python
1 - ((1-r2)*(40-1)/(40-1-1))
```
